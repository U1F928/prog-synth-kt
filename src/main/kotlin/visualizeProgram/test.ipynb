```kotlin
%use lets-plot

%use lets-plot

// Define nodes and edges
        data class Node(val id: String, val label: String)
data class Edge(val source: String, val target: String)

val nodes = listOf(
    Node("a", "Node A"),
    Node("b", "Node B"),
    Node("c", "Node C"),
    Node("d", "Node D")
)

val edges = listOf(
    Edge("a", "b"),
    Edge("b", "c"),
    Edge("c", "d"),
    Edge("d", "a")
)

// Compute circular layout
val centerX = 0.0
val centerY = 0.0
val radius = 1.0
val angleStep = 2 * Math.PI / nodes.size

val positions = nodes.mapIndexed { i, node ->
    val angle = i * angleStep
    node.id to Pair(radius * kotlin.math.cos(angle), radius * kotlin.math.sin(angle))
}.toMap()

// Prepare data for edges (segments)
val edgeData = mapOf(
    "x" to edges.map { positions[it.source]!!.first },
    "y" to edges.map { positions[it.source]!!.second },
    "xend" to edges.map { positions[it.target]!!.first },
    "yend" to edges.map { positions[it.target]!!.second }
)

// Prepare data for nodes
val nodeData = mapOf(
    "x" to nodes.map { positions[it.id]!!.first },
    "y" to nodes.map { positions[it.id]!!.second },
    "label" to nodes.map { it.label }
)

// Build the plot
val plot = letsPlot() +
        geomSegment(edgeData ) +
        geomPoint(nodeData) +
        geomText(nodeData)

plot
```
